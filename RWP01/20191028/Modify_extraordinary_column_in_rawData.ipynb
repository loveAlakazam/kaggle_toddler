# 실제 데이터의 컬럼

```python
sample_seoul_data_target_columns = sample_seoul_data.iloc[15].values
sample_seoul_data_target_columns
```

# 컬럼명 바꾸기

```python
changed_cols=dict([(org,new) for org, new in zip(sample_seoul_data.columns, sample_seoul_data_target_columns)])
changed_cols
```

```python
# 컬럼 바꾸기 
sample_seoul_data.rename( columns=changed_cols, inplace=True)
```
