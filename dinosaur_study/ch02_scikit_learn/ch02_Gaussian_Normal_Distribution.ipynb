# 표준정규분포 구현하기

표준 정규 분포(Standard Normal Distribution) 정의

- 평균(m)= 0
- 분산(σ**2)=1

[참고 자료]<br> https://datascienceschool.net/view-notebook/76acc92d28354e86940001f9fe85c50f/ 

```python
import scipy as sp
import scipy.stats
```

```python
m =0 #평균
std = 1 #표준편차=1, 분산=(표준편차)**2=1
rv = sp.stats.norm(m, std) 
```

```python
# pdf 메소드를 사용하여 확률밀도 함수를 계산할 수 있다.
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm
fontprop=fm.FontProperties(fname='malgun.ttf', size=15)

x_axis = np.linspace(-10, 10, 100)
plt.grid(alpha=0.5)
plt.plot(x_axis, rv.pdf(x_axis))
plt.ylabel('확률 밀도값 p(x)', fontproperties=fontprop)
plt.title('표준 정규 분포의 확률 밀도함수',fontproperties=fontprop)
plt.show()
```

# 랜덤 표본 생성시켜서 시각화

```python
np.random.seed(0)
x= rv.rvs(20)  # 표본 20개를 얻는다. rvs() : 표본을 얻는 메소드
x
```

```python
import seaborn as sns
sns.distplot(x, kde=True, fit=sp.stats.norm, color='pink')
plt.grid(alpha=0.5)
plt.title('랜덤 표본 생성결과', fontproperties=fontprop)
plt.xlabel('표본값',fontproperties=fontprop)
plt.ylabel("$p(x)$")
plt.show()
```
